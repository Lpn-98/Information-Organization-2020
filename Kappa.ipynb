# Kappa

输入Judge1和Judge2的结果，输出混淆矩阵和Kappa

```python
print('输入Judge1的名称')
name_j1=input()

print('输入Judge1的结果')
judge1=[]
while True:
    input_=input()
    judge1.append(input_)
    if not input_ :
        break
    if input_!='N' and input_!='R':
        print("输入了‘N’和‘R’以外的内容，请检查！")
        judge1=[]
        break

print('输入Judge2的名称')
name_j2=input()

print('输入Judge2的结果')
judge2=[]
while True:
    input_=input()
    judge2.append(input_)
    if not input_:
        break
    if input_!='N' and input_!='R':
        print("输入了‘N’和‘R’以外的内容，请检查！")
        judge2=[]
        break

if len(judge1)!=len(judge2):
    print("Judge1和Judge2长度不同！")
else:
    RR=sum([1 for i in range(len(judge1)) if judge1[i]=='R' and judge2[i]=='R'])
    RN=sum([1 for i in range(len(judge1)) if judge1[i]=='R' and judge2[i]=='N'])
    NR=sum([1 for i in range(len(judge1)) if judge1[i]=='N' and judge2[i]=='R'])
    NN=sum([1 for i in range(len(judge1)) if judge1[i]=='N' and judge2[i]=='N'])
    print("【矩阵：】")
    print('             ,{} , {} , {}'.format(name_j2+' R ',name_j2+' N ',name_j2+' Total '))
    print('{}     ,  {}  ,  {}  ,  {}'.format(name_j1+' R ',RR,RN,RR+RN))
    print('{}     ,  {}  ,  {}  ,  {}'.format(name_j1+' N ',NR,NN,NR+NN))
    print('{} ,  {}  ,  {}  ,  {}'.format(name_j1+' Total ',RR+NR,RN+NN,RR+RN+NR+NN))

T=round(RR+RN+NR+NN,2)
PA=round((RR+NN)/T,2)
print('P(A)=({}+{})/{}={}'.format(RR,NN,T,PA))
PN=round((NR+NN+RN+NN)/(T+T),2)
print('P(nonrelevant)=({}+{})/({}+{})={}'.format(NR+NN,RN+NN,T,T,PN))
PR=round((RR+RN+RR+NR)/(T+T),2)
print('P(relevant)=({}+{})/({}+{})={}'.format(RR+RN,RR+NR,T,T,PR))
print('Probability that the two judges aggreed by chance:')
PE=round(PN**2+PR**2,2)
print('P(E)=P(nonrelevant)^2+P(relevant)^2={}^2+{}^2={}'.format(PN,PR,PE))
print("Kappa statistic:")
kappa=round((PA-PE)/(1-PE),2)
print('kappa=(P(A)-P(E))/(1-P(E))=({}-{})/(1-{})={}'.format(PA,PE,PE,kappa))
if kappa > 2/3:
    print('acceptable')
else:
    print('【not acceptable】')
```
